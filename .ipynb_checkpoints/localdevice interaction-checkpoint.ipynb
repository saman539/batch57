```python
def kill_process(process_name):
    for row in resp.splitlines():
        if process_name in row.decode("utf-8"):
            pid = row.split()[1]
            pid=int(pid)
            os.system("kill %s"%pid)
            return True
    return False
process_name = input("enter process name:")
result = kill_process(process_name)
if result:
    print("process killed successfully")
else:
    print("issue in the killing")
```
